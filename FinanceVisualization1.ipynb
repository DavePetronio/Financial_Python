```python
#Author: David Petronio
#This notebook reads a financial history CSV file in order to perform calculations and display relevant data
#import pandas, use it to read a user-inputted CSV file
#Try/Except error checking prints if file name is invalid
import pandas as pd
try:
    csv_file = input("Please enter the CSV file name: ")
    fh = pd.read_csv(csv_file)
except:
    print("An error has occured, please make sure the filename and path are entered correctly.")
```

```python
#display as percent
daily_pct_change = (fh['Close'] - fh['Open'])/fh['Open'] * 100
daily_pct_change
```

```python
#add calculated attribute daily change
fh['Daily_Change'] = fh['Close'] - fh['Open']
```

```python
#add calculated attribute daily gain
fh['Daily_Gain'] = fh['Close'] / fh['Close'].iloc[0]
```
